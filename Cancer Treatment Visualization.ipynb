```python
#Dependencies
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
import numpy as np
```

```python
#Import csv files
mouse_data = pd.read_csv("./Mouse_metadata.csv")
study_results = pd.read_csv("./Study_results.csv")

#Merge DataFrames
merged_df = pd.merge(study_results, mouse_data, on="Mouse ID", how="left")
merged_df
```

```python
#Show number of mice
unique_mice = len(merged_df["Mouse ID"].unique())
unique_mice
```

```python
#Check for any Mouse ID with duplicate timepoints
duplicated_mouse_ids = merged_df[merged_df.duplicated(subset = ["Mouse ID", "Timepoint"])]["Mouse ID"].unique()
duplicated_mouse_ids
```

```python
#Create a new DataFrame with duplicates removed
clean_complete_df = merged_df[merged_df["Mouse ID"].isin(duplicated_mouse_ids) == False]
clean_complete_df
```

```python
#Show number of Mouse IDs after duplicates were dropped
print(len(clean_complete_df["Mouse ID"].unique()))
```

```python
#Set variables to add as columns in new DataFrame
tumor_volume_mean = clean_complete_df.groupby(["Drug Regimen"]).mean()["Tumor Volume (mm3)"]
tumor_volume_median = clean_complete_df.groupby(["Drug Regimen"]).median()["Tumor Volume (mm3)"]
tumor_volume_variance = clean_complete_df.groupby(["Drug Regimen"]).var()["Tumor Volume (mm3)"]
tumor_volume_stdev = clean_complete_df.groupby(["Drug Regimen"]).agg(np.std)["Tumor Volume (mm3)"]
tumor_volume_sem = clean_complete_df.groupby(["Drug Regimen"]).sem()["Tumor Volume (mm3)"]
```

```python
#Create summary statistics DataFrame
summary_tumor_volume = pd.DataFrame({
   
    "Mean Tumor Volume": tumor_volume_mean,
    "Median Tumor Volume": tumor_volume_median,
    "Tumor Volume Variance": tumor_volume_variance,
    "Tumor Volume Std. Dev.": tumor_volume_stdev,
    "Tumor Volume SEM": tumor_volume_sem
    
})

#Display
summary_tumor_volume
```

```python
#Create a DataFrame containing the last timepoint for each mouse
last_timepoint_df = clean_complete_df.groupby(["Mouse ID"])["Timepoint"].max()
last_timepoint_df.reset_index()
```

```python
#Merge to create a final DataFrame to use for charting
merged_final_timepoint_df = pd.merge(last_timepoint_df, clean_complete_df, on=["Mouse ID", "Timepoint"])
merged_final_timepoint_df.head(20)
```

```python
#Create lists
treatment_list = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]
tumor_volume_data = []

#Use a for loop to get the final tumor volume of each mouse in the 4 treatment regimens listed above
for drug in treatment_list:
    final_tumor_vol = merged_final_timepoint_df.loc[merged_final_timepoint_df["Drug Regimen"] == drug, "Tumor Volume (mm3)"]
    tumor_volume_data.append(final_tumor_vol)

#Calculate outliers    
    quartiles = final_tumor_vol.quantile([.25, .5, .75])
    lowerq = quartiles[.25]
    upperq = quartiles[.75]
    iqr = upperq - lowerq
    lower_bound = lowerq - (1.5*iqr)
    upper_bound = upperq + (1.5*iqr)

    outliers = final_tumor_vol.loc[(final_tumor_vol < lower_bound) | (final_tumor_vol > upper_bound)]
    print(f"{drug}'s potential outliers: {outliers}")
    
```

```python
#Create a boxplot to show final tumor volume results across the 4 treatment regimens & their outliers
color = dict(markerfacecolor = "red", markersize = 8)
plt.boxplot(tumor_volume_data, labels=treatment_list, flierprops = color)
plt.ylabel("Final Tumor Volume (mm3)")
plt.show()
```
